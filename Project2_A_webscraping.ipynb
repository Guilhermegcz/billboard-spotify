### coletando dados

```python
df = pd.DataFrame()

for ano in range(2006,2020):
    
    #condicao para procurar por ano
    
    url = (f'https://www.billboard.com/charts/year-end/{ano}/hot-100-songs')
    response = requests.get(url)
    html = response.content
    soup = BeautifulSoup(html)

    for position in range(100):
        try:
            musica = soup.find_all('div', attrs={'class':'ye-chart-item__title'})[position].contents[0].strip()

            #condiÃ§Ã£o para artista
            if soup.find_all('div', attrs={'class':'ye-chart-item__artist'})[position].contents[0].strip() == '':
                artista = soup.find_all('div', attrs={'class':'ye-chart-item__artist'})[position].contents[1].contents[0].strip()
            else: 
                artista = soup.find_all('div', attrs={'class':'ye-chart-item__artist'})[position].contents[0].strip()

            foto = soup.find_all('div', attrs={'class':'ye-chart-item__image'})[position].contents[1].get('src')
        except:
            pass
    
        mini_df = pd.DataFrame({'rank' : [position + 1],
                            'music' : [musica],
                            'artist': [artista],
                            'photo': [foto],
                            'year' : [ano]
                               })
        df = pd.concat([df, mini_df])   
```
