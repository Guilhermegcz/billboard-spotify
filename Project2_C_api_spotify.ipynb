# função para verificar a musica e o artista

```python
# funcao para verificar se a musica bate com o artista selecionado
def getting_info(music):
    ''' This function collect information from spotify'''

    music_list_found = sp.search(q='track:' + music, type='track')
    return music_list_found
```

# função para coletar o id da musica

```python
id_dict = {}
def check_track(dataframe, column_music, column_artist, colum_id):
    '''From a dataframe this function will your music id'''
    
    for index in range(dataframe[column_music].size):
        music = dataframe[column_music][index]
        artist = dataframe[column_artist][index]
        print(index)
        
        music_found = getting_info(music)
        #o spotify retorna sempre 10 resultados na primeira pagina
        
        for position in range(10):
            #acessando cada resultado
            try:
                music_in_list = music_found['tracks']['items'][position]['name']
                artist_in_list = music_found['tracks']['items'][position]['artists'][0]['name']


                #removendo os apostrofes
                music_in_list = music_in_list.replace("'", "").replace("’","")
                artist_in_list = artist_in_list.replace("'", "").replace("’","")
                music = music.replace("'", "").replace("’","")
                artist = artist.replace("'", "").replace("’","")


                #removendo os acentos
                music_in_list = unidecode.unidecode(music_in_list)
                artist_in_list = unidecode.unidecode(artist_in_list)
                music = unidecode.unidecode(music)
                artist = unidecode.unidecode(artist)


                #verificar se a musica da lista bate com a musica do resultado
                if re.findall(f'{music}', music_in_list) != [] and re.findall(f'{artist}', artist_in_list) != []:

                    id_music = music_found['tracks']['items'][position]['id'] 
                    id_dict[index] = id_music
                    #print("------------------")
                    #print(music_found['tracks']['items'][position]['id'])
                    print(music_found['tracks']['items'][position]['artists'][0]['name'])
                    #print(music_found['tracks']['items'][position]['name'])
                    #print("------------------")
                    break
                else:
                    pass
                    #print('-------nao combina------------')
                    #print(music_in_list)
                    #print(music)
                    #print(artist_in_list)
                    #print(artist)
            except:
                pass
```

```python
# função para buscar e armazenar mais informação sobre a musica
def collecting_data_music(music_id_colum, info):
    try:
        return_features = sp.audio_features(music_id_colum)
        return return_features[0][info]
    except:
        pass
```

```python
test_capt['id_music']
```

```python
test_capt['id_music'].apply(collecting_data_music, args=['danceability'])
```

```python
def df_creating_columns(dataframe, list_columns):
    try:
        for infos in list_columns:
            dataframe[infos] = dataframe['id_music'].apply(collecting_data_music, args=[infos])
    
    except:
        pass   
    
```

```python
infos = ['danceability', 'energy', 'loudness','speechiness','acousticness','instrumentalness','tempo','uri','duration_ms']
```
