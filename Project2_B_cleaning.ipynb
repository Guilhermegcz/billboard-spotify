# LOADING DATA

```python
path = 'C:/Users/dell/IRONHACK/Ironhack-project/Project2_data-gathering'
df_bil = pd.read_csv(path + '/top_billboard.csv')
```

# CLEANING DATA

```python
df_bil['main_artist'] = 0
df_bil['featuring1'] = 0
df_bil['featuring2'] = 0
```

```python
#função para separar os artistas que fizeram featuring
def split_artist(artist):
    '''This function split the artist from your featurings.'''
    
    lista_artist = re.split(' & | Featuring ', artist)
    
    return lista_artist
```

```python
df_bil['list_artist'] = df_bil['artist'].apply(split_artist)
```

```python
#função para pegar o primeiro artista
def main_artist(artist):
    main = artist[0]
    return main
```

```python
df_bil['main_artist'] = df_bil['list_artist'].apply(main_artist)
```

```python
#função para pegar featuring
def featuring(artist,num_featuring=1):
    try:
        feat = artist[num_featuring]
        return feat
    except:
        pass
```

```python
df_bil['featuring1'] = df_bil['list_artist'].apply(featuring)
```

```python
df_bil['featuring2'] = df_bil['list_artist'].apply(featuring,args=[2])
```

```python
df_bil.head()
```

```python
year_billboard = df_bil.drop(['artist', 'list_artist'],axis=1)
```

```python
year_billboard = year_billboard[['rank','music','main_artist','year','featuring1','featuring2','photo']]
```

### EXPORTANDO OS DADOS
